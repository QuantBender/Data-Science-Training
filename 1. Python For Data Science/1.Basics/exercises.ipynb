````python
# TODO: Edit the function to return the correct bill for different
# values of num_gallons
def get_water_bill(num_gallons):
    bill = 1000
    if(num_gallons <= 8000):
        bill = (num_gallons / 1000) * 5
    elif(num_gallons >= 8001 and num_gallons <= 22000):
        bill = (num_gallons / 1000) * 6
    elif(num_gallons >= 22001 and num_gallons <= 30000):
        bill = (num_gallons / 1000) * 7
    else:
        bill = (num_gallons / 1000) * 10
    return bill
````

````python
get_water_bill(25000)
````

````python
# TODO: Edit the function to return the correct bill for different
# values of GB
def get_phone_bill(gb):
    bill = 100
    if(gb <= 15):
        return bill
    else:
        over_gb = gb - 15
        over_bill = over_gb * 100
        real_bill = bill + over_bill
    return real_bill
````

````python
get_phone_bill(15.1)
````
